# Matplotlib Nedir? 

--> Matplotlib, görselleştirme aracı olarak hizmet veren python'da düşük seviyeli bir grafik çizim kitaplığıdır.

## Matplotlib Kütüphanesini Çağırma

```python
import matplotlib.pyplot as plt
```

## Grafikler ve Grafiklerde İşlemler

###### Her işlemin sonunda çizilen grafiği görüntülemek için yazılması gereken komut plt.show() 

```python
plt.show()
```

###### Grafiğe başlık ekleme --> plt.title()

```python
plt.title("Yalın çizdirim")
```

###### X ve Y eksenlerine isim verme --> plt.xlabel() ve plt.ylabel() 

```python
plt.xlabel("X Değerleri") # X eksenine isim verir
plt.ylabel("Y Değerleri") # Y eksenine isim verir
```

###### Çizgi / Trend grafiği çizdirme --> plt.plot() 

--> Çizgi grafiğinde:
X ekseninde zaman, Y ekseninde zamana karşılık gelen değerler yer alır.

```python
year = [1950, 1970, 1990, 2010]
pop = [2519, 3692, 5263, 6972]
```

```python
plt.plot(year, pop) # plt.plot(x ekseni, y ekseni)
```

###### Dağılım grafiği çizdirme --> plt.scatter() 

```python
plt.scatter(year, pop) # plt.scatter(x ekseni, y ekseni)
```

###### X ve Y eksenlerinin ölçeğini belirleme --> plt.xscale() ve plt.yscale() 

```python
plt.scatter(year, pop)
plt.xscale("log") # X ekseninin ölçeğini belirler
plt.yscale("log") # Y ekseninin ölçeğini belirler
```

###### Histogram grafiği çizdirme --> plt.hist()

 --> Histogram grafiği:
Sürekli nicel veriler için uygun bir grafik türüdür. 

```python
life_exp = [2.12, 3.42, 2.76, 3.18, 0.50, 2.14, 2.35, 3.70, 2.84, 1.96, 1.29, 2.36, 2.78, 2.80, 2.15, 1.65, 1.03, 3.87, 2.90] 
```

```python
plt.hist(life_exp, bins = None, range = None)
```

--> plt.hist(bins = verilerin kaç kutuya bölünmesi gerektiğini söyleyen parametredir)

```python
plt.hist(life_exp, bins = 20) 
```

###### Mevcut rakamı temizlemek plt.clf() ile yapılır. Bir grafiği çizdirdikten sonra yeni bir grafik daha çizdirmeden önce yazılır. 

```python
plt.clf()
```

###### X ve Y eksenlerinin sayı aralıklarını düzenleme --> plt.yticks([....]) ve plt.xticks([....])

```python
lst_1 = [23,34,45,67,78,90,103,74]
lst_2 = [96,96,96,78,84,27,2,45]
```

```python
plt.plot(lst_1, lst_2)
```

--> Eksenlerdeki sayılar plt.xticks ve plt.yticks'teki listelerdeki sayılara dönüşür.

```python
plt.yticks([0,15,30,45,60,75,90]) 
plt.xticks([30, 60, 90])
plt.plot(lst_1, lst_2)
```

--> Eksenlerdeki sayıların neyi belirttiğini daha düzgün ifade edebilmek için aşağıda olduğu gibi plt.x,yticks() içerisinde eksendeki sayılarla aynı sayıda olacak şekilde bir liste daha oluşturup belirtebiliriz.  

```python
plt.yticks([0,15,30,45,60,75,90],
          ["0", "15M", "30M", "45M", "60M", "75M", "90M"])
plt.xticks([0, 30, 60, 90],
          ["0", "30K", "60K", "90K"])
plt.plot(lst_1, lst_2)
```

###### Tabloyu daha iyi takip edebilmek için tablonun arkasına grid (kafes) eklemek --> plt.grid(True) 

```python
plt.grid(True)
plt.xlabel("lst_1")
plt.ylabel("lst_2")
plt.title("Gridli tablo")

plt.plot(lst_1, lst_2)
plt.show()
```

###### Bar grafiği çizdirme --> plt.bar() 

--> Bar grafiği:
Nitel veriler için uygun bir grafik türüdür. 

```python
names = ["Java", "C#", "Python", "PHP"]
values = [3, 7, 2, 9]

plt.bar(names, values, color = "red")
plt.show
```

###### Bar grafiğinin değerlerini Dictionary (sözlük)'ten almak 

```python
dictionary = {}
dictionary["languages"] = names
dictionary["values"] = values

plt.bar(names, values, data = dictionary)
plt.show()
```

###### Pasta grafiği çizdirme --> plt.pie() 

--> Pasta grafiği:
Nitel veriler için uygun bir grafik türüdür.

```python
plt.pie(values, labels = names, autopct = "%1.2f%%") #lables = Eksenler / values = Veriler / autopct = Yüzdelik Değerler
```

###### Boxplot çizdirme --> plt.boxplot() 

--> Aykırı değer incelemelerinde boxplot (kutu grafiği) kullanılır.

```python
import numpy as np
lst_3 = np.random.normal(size = 100)
lst_4 = np.random.uniform(size = 100)
data = [lst_3, lst_4]
```

```python
plt.boxplot(data)
```
